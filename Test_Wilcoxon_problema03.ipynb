# Código donde se ejecuta el test de Wilcoxon para los resultados

Evaluaremos el baseline (workflow sin modificaciones y sin Feature engineering manual) vs ambos experimentos realizados.

```r
# Ganancias del baseline
ganancias_baseline <- c(5918400, 6268200, 5664000, 5756100, 5865001)

# Ganancias del experimento con variables financieras
ganancias_experimento1 <- c(5851200, 5829000, 5470200, 5645700, 5865000)

# Test de Wilcoxon para verificar si baseline es mejor que experimento
resultado <- wilcox.test(ganancias_baseline, ganancias_experimento1, paired = TRUE, alternative = "greater")

print(resultado)
```

```r
# Ganancias del baseline
ganancias_baseline <- c(5918400, 6268200, 5664000, 5756100, 5865000)

# Ganancias del experimento con variables financieras combinadas con demográficas
ganancias_experimento2 <- c(5697900, 5684700, 5606700, 5574000, 5807400)

# Test de Wilcoxon para verificar si baseline es mejor que experimento
resultado <- wilcox.test(ganancias_baseline, ganancias_experimento2, paired = TRUE, alternative = "greater")

print(resultado)
```
